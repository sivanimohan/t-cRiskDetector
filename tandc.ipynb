```python
import os

extract_path = "/content/drive/MyDrive/text"  # Change if needed
txt_files = [f for f in os.listdir(extract_path) if f.endswith(".txt")]

print("Extracted TXT files:", txt_files)
```

```python
import pandas as pd

data = []

for file in txt_files:
    file_path = os.path.join(extract_path, file)
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()
        data.append((file, content))  # Store filename and content

df = pd.DataFrame(data, columns=["filename", "text"])



print(df.head())
```

```python
import re

def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    return text

df["clean_text"] = df["text"].apply(preprocess_text)
print(df.head())


```

```python
import torch
from transformers import pipeline

# Load BERT model for text classification
classifier = pipeline("text-classification", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Risk category mapping
risk_labels = {0: "Low", 1: "Medium", 2: "High"}

# Function to classify risk using BERT
def classify_risk(text):
    result = classifier(text[:512])[0]  # Truncate long text to 512 tokens
    label = result["label"]

    if "1 star" in label or "2 star" in label:
        return 0  # Low risk
    elif "3 star" in label:
        return 1  # Medium risk
    else:
        return 2  # High risk

# Function to analyze and explain risk
def analyze_risk(text):
    sentences = text.split(". ")  # Split into sentences
    high_risk_sentences = []
    moderate_risk_sentences = []
    low_risk_sentences = []

    total_risk_score = 0

    for sentence in sentences:
        risk = classify_risk(sentence)
        total_risk_score += risk

        if risk == 2:
            high_risk_sentences.append(sentence)
        elif risk == 1:
            moderate_risk_sentences.append(sentence)
        else:
            low_risk_sentences.append(sentence)

    avg_risk_score = total_risk_score / max(1, len(sentences))  # Avoid division by zero
    risk_level = risk_labels[round(avg_risk_score)]  # Assign risk level

    # Create risk explanation
    explanation = f"This document has an overall {risk_level} risk level."
    if high_risk_sentences:
        explanation += f" High-risk sections include: {', '.join(high_risk_sentences[:3])}..."
    if moderate_risk_sentences:
        explanation += f" Moderate-risk sections include: {', '.join(moderate_risk_sentences[:3])}..."
    if low_risk_sentences:
        explanation += f" Some low-risk sections: {', '.join(low_risk_sentences[:3])}..."

    return avg_risk_score, risk_level, explanation

# Apply risk analysis to each document
df["risk_score"], df["risk_level"], df["risk_explanation"] = zip(*df["clean_text"].apply(analyze_risk))

# Show results
print(df[["filename", "risk_score", "risk_level", "risk_explanation"]].head())
```

```python
import os
from transformers import AutoTokenizer, TrainingArguments, Trainer

# Disable WandB logging
os.environ["WANDB_DISABLED"] = "true"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("nlpaueb/legal-bert-base-uncased")

# Tokenize function
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Apply tokenization
tokenized_dataset = dataset.map(tokenize_function, batched=True)
tokenized_dataset = tokenized_dataset.remove_columns(["text"])

# Split into train & eval
tokenized_dataset = tokenized_dataset.train_test_split(test_size=0.2)
train_dataset = tokenized_dataset["train"]
eval_dataset = tokenized_dataset["test"]

# Ensure dataset format
train_dataset.set_format("torch")
eval_dataset.set_format("torch")

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",  # ✅ Updated argument
    save_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    logging_dir="./logs",
    fp16=True,
    logging_steps=500,
    report_to="none"  # ✅ Disable W&B logging properly
)

# Initialize Trainer (removed deprecated tokenizer argument)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)

# Train the model
trainer.train()
```
